#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(s: str) -> int:
    """
    Считает количество гласных в строке.
    Учитываются только латинские буквы (a, e, i, o, u), регистр не важен.
    """
    vowels = {'a', 'e', 'i', 'o', 'u'}
    return sum(ch in vowels for ch in s.lower())


def test_count_vowels():
    # базовые проверки
    assert count_vowels('hello') == 2
    assert count_vowels('rhythm') == 0
    assert count_vowels('AEIOU') == 5

    # дополнительные случаи
    assert count_vowels('abcdefghijklmnopqrstuvwxyz') == 5
    assert count_vowels('') == 0          # пустая строка
    assert count_vowels('12345!') == 0    # нет букв
    assert count_vowels('A quick brown fox.') == 5

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_count_vowels()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def has_unique_chars(s: str) -> bool:
    """
    Проверяет, что все символы в строке уникальны.
    Возвращает True, если повторов нет, иначе False.
    """
    return len(set(s)) == len(s)


def test_has_unique_chars():
    # базовые проверки
    assert has_unique_chars('abc') is True
    assert has_unique_chars('aabb') is False
    assert has_unique_chars('') is True          # пустая строка — уникальна
    assert has_unique_chars('Aa') is True        # учитывается регистр
    assert has_unique_chars('Hello') is False    # есть повтор 'l'

    # дополнительные случаи
    assert has_unique_chars('123!@#') is True
    assert has_unique_chars('1123') is False
    assert has_unique_chars('Python') is True

    print("Все тесты прошли успешно!")


if __name__ == "__main__":
    test_has_unique_chars()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_ones(n):
    if n <= 0:
        raise ValueError("Число должно быть положительным")
    count = 0
    while n > 0:
        count += n & 1  
        n >>= 1         
    return count
#Тесты
def test_count_ones():
    assert count_ones(1) == 1       
    assert count_ones(2) == 1       
    assert count_ones(3) == 2      
    assert count_ones(7) == 3      
    assert count_ones(15) == 4      
    assert count_ones(1023) == 10   
    try:
        count_ones(0)
        assert False, 
    except ValueError:
        pass
    try:
        count_ones(-5)
        assert False, 
    except ValueError:
        pass

    print("Все тесты пройдены!")

test_count_ones()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def mult_numbers(n: int) -> int:
    res = 1
    for num in str(n):
        res *= int(num)
    return res

def multiplicative_persistence(n: int) -> int:
    if n <= 0:
        raise ValueError("Число должно быть положительным")
    count = 0
    while n >= 10:
        n = mult_numbers(n)
        count += 1
    return count

def test_multiplicative_persistence():
    assert multiplicative_persistence(39) == 3
    assert multiplicative_persistence(4) == 0
    assert multiplicative_persistence(999) == 4
    assert multiplicative_persistence(10) == 1
    try:
        multiplicative_persistence(0)
        assert False
    except ValueError:
        pass
    try:
        multiplicative_persistence(-5)
        assert False
    except ValueError:
        pass
    print("Все тесты пройдены!")

if __name__ == "__main__":
    test_multiplicative_persistence()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mean_squared_error(vec1, vec2):
    if len(vec1) != len(vec2):
        raise ValueError("Векторы должны быть одинаковой длины")
    return sum([(a - b) ** 2 for a, b in zip(vec1, vec2)]) / len(vec1)

def test_mean_squared_error():
    assert mean_squared_error([1, 2, 3], [1, 2, 3]) == 0
    assert mean_squared_error([1, 2, 3], [2, 3, 4]) == 1
    assert mean_squared_error([0, 0, 0], [1, 1, 1]) == 1
    assert mean_squared_error([1, 2], [3, 4]) == 4
    assert round(mean_squared_error([1, 2, 3], [1, 3, 5]), 2) == 1.67
    try:
        mean_squared_error([1, 2], [1, 2, 3])
        assert False
    except ValueError:
        pass
    print("Все тесты пройдены!")

if __name__ == "__main__":
    test_mean_squared_error()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors_string(n: int) -> str:
    if n <= 0:
        raise ValueError("Число должно быть положительным")
    
    result = []
    divisor = 2
    while n > 1:
        count = 0
        while n % divisor == 0:
            n //= divisor
            count += 1
        if count > 0:
            if count == 1:
                result.append(f"({divisor})")
            else:
                result.append(f"({divisor}**{count})")
        divisor += 1
        if divisor * divisor > n and n > 1:
            result.append(f"({n})")
            break
    return ''.join(result)


#Тесты 
def test_prime_factors_string():
    assert prime_factors_string(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factors_string(60) == "(2**2)(3)(5)"
    assert prime_factors_string(13) == "(13)"
    assert prime_factors_string(100) == "(2**2)(5**2)"
    try:
        prime_factors_string(0)
        assert False
    except ValueError:
        pass
    try:
        prime_factors_string(-5)
        assert False
    except ValueError:
        pass
    print("Все тесты пройдены!")

if __name__ == "__main__":
    test_prime_factors_string()
  
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid_height(n: int):
    if n <= 0:
        return "It is impossible"
    k = 0
    total = 0
    while total < n:
        k += 1
        total += k ** 2
    return k if total == n else "It is impossible"


#Тесты 
def test_pyramid_height():
    assert pyramid_height(14) == 3
    assert pyramid_height(15) == "It is impossible"
    print("Все тесты пройдены!")


if __name__ == "__main__":
    test_pyramid_height()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced_number(n: int) -> bool:
    if n <= 0:
        raise ValueError("Число должно быть положительным")
    
    digits = [int(d) for d in str(n)]
    length = len(digits)
    
    if length == 1:
        return True
    
    mid = length // 2
    if length % 2 == 0:
        left = digits[:mid-1]
        right = digits[mid+1:]
    else:
        left = digits[:mid]
        right = digits[mid+1:]
    
    return sum(left) == sum(right)


#Тесты 
def test_is_balanced_number():
    assert is_balanced_number(123803) == True 
    assert is_balanced_number(12345) == False
    assert is_balanced_number(7) == True
    print("Все тесты пройдены!")

if __name__ == "__main__":
    test_is_balanced_number()
```
